```python
import pandas as pd
import gzip

def read_gzip_file(file_path):
    """
    Read a .tsv.gz compressed dataset file into a pandas DataFrame.

    Parameters:
    file_path (str): The file path to the .tsv.gz file.

    Returns:
    pd.DataFrame: A DataFrame containing the data from the .tsv.gz file.
    """
    # Use gzip.open to read the gzipped file in binary mode ('rb'), then parse it with pandas
    with gzip.open(file_path, 'rb') as file:
        df = pd.read_csv(file, sep='\t')

    return df
```

```python
import gzip
import shutil

def write_to_gzipfile(tsv_filename, tsv_gz_filename):
    with open(tsv_filename, 'rb') as f_in:
        with gzip.open(tsv_gz_filename, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```

```python
import pandas as pd

gz_file_extension = '.gz'
tsv_filename = 'main.tsv'
tsv_gz_filename = tsv_filename + gz_file_extension; 
tsv_gz_file_path = tsv_gz_filename;

df = read_gzip_file(tsv_gz_file_path)

# get subset from the dataframe
subset = df.loc[:20000];

tsv_sample_filename = 'subset_sample.tsv';

# convert to tsv file
subset.to_csv(tsv_sample_filename, sep="\t")

tsv_gz_sample_filename = tsv_sample_filename + gz_file_extension;

write_to_gzipfile(tsv_sample_filename, tsv_gz_sample_filename)
```

```python

def dcg(relevances, rank=10):
    """Discounted cumulative gain at rank (DCG)"""
    relevances = np.asarray(relevances)[:rank]
    n_relevances = len(relevances)
    if n_relevances == 0:
        return 0.

    discounts = np.log2(np.arange(n_relevances) + 2)
    return np.sum(relevances / discounts)
 
 
def ndcg(relevances, rank=10):
    """Normalized discounted cumulative gain (NDGC)"""
    best_dcg = dcg(sorted(relevances, reverse=True), rank)
    if best_dcg == 0:
        return 0.

    return dcg(relevances, rank) / best_dcg
```
